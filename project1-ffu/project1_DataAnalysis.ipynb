```python
#plot the number of columns to be dropped as a function of the coefficient of variation CV (see definition below)
cv_min_list = [1e-4,1e-3,1e-2,2e-2,3e-2,4e-2,5e-2,6e-2,7e-2,8e-2,9e-2,1e-1]
n_dropped_list = []
n=0
for cv_min in cv_min_list:
    for column in df_x:
        mean = df_x[column].astype("float").mean()
        if mean == 0:
           n = n+1
        else: 
            cv = df_x[column].astype("float").std()/mean
            if cv < cv_min:
              n = n+1
    print("CV min:",cv_min, "Features to be dropped:", n)          
    n_dropped_list.append(n)
    n=0
```

```python
#identify columns which contains constant values and some noise.
#use the coefficient of variation CV as metric (CV = sigma/mean).
#drop columns in which CV < 1e3 (try different values of CV)
#I have checked out that cv_min < 1e-2 has no impact. However, choosing cv_min < 1e-1 removes additionally ~600 columns. To be checked
cv_min = 1e-1
dropped_cols_mean = [] #zero mean
dropped_cols_cv = [] #CV
n = 0
n_tot = len(df_x.columns)
for column in df_x:
    mean = df_x[column].astype("float").mean()
    if mean == 0:
        print("mean = 0: dropping column", column)
        dropped_cols_mean.append(column)
        n = n+1
    else: 
        cv = df_x[column].astype("float").std()/mean
        if cv < cv_min:
            print("cv < ", cv_min,": dropping column", column)
            dropped_cols_cv.append(column)
            n = n+1
#print("dropped columns %d / %d",n,n_tot)
print(n,n_tot)
```

```python
#check that indeed the column contains only zeros
df_x["x629"]
```

```python
#drop the columns with mean = 0 and cv < cv_min
df_x.drop(dropped_cols_mean,axis=1,inplace=True)
df_x.drop(dropped_cols_cv,axis=1,inplace=True)
```

### Ideas
1. group the features by their pearson correlation coefficient (with respect to y)
  * 1 positive correlation
  * 0 no correlation
  * 1 negative correlation
2. check the statistical significance with p value
  * p-value is $<$ 0.001: we say there is strong evidence that the correlation is significant
  * the p-value is $<$ 0.05: there is moderate evidence that the correlation is significant
  * the p-value is $<$ 0.1: there is weak evidence that the correlation is significant
  * the p-value is $>$ 0.1: there is no evidence that the correlation is significant

```python
#create df from iso by removing the target
iso_Y_arr = df_top["y"].to_numpy()
df_iso = df_top.drop(["y"],axis=1)
iso_X_arr = df_iso.to_numpy()
print(iso_X_arr.shape)
#df_iso.head(5)
```

```python
# use Isolation Forest to indentify outliers on the selected features
from sklearn.ensemble import IsolationForest
#cont_fl = 0.3 #use auto --> specifying a value is interpreted as % of outliers
iso = IsolationForest() 
yhat = iso.fit_predict(iso_X_arr)
print(yhat.shape, iso_X_arr.shape)
mask = yhat != -1
iso_X_arr, iso_Y_arr = iso_X_arr[mask, :], iso_Y_arr[mask]
numrows = len(iso_X_arr)    
numcols = len(iso_X_arr[0])
iso_arr = np.random.rand(numrows,numcols+1)
iso_arr[:,0]=iso_Y_arr
iso_arr[:,1:(numcols+1)]=iso_X_arr
print(numrows,numcols)
print(iso_arr.shape)
```

```python
# compare against plots before Isolation forest
df_top_iso = pd.DataFrame(data=iso_arr, columns=df_top_colnames)
df_top_iso.head()
```
